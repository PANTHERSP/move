```python
import numpy as np

# กำหนดโครงสร้างข้อมูล (structured dtype) ที่มีฟิลด์ name และ age
data_type = np.dtype([('name', np.str_, 16), ('age', np.int32)])

# สร้าง Array scalar แบบ structure โดยกำหนดค่าให้กับแต่ละฟิลด์
person = np.array(('John', 30), dtype=data_type)

# แสดงข้อมูลใน Array scalar
print(person)  # Output: ('John', 30)
print(person['name'])  # Output: 'John'
print(person['age'])   # Output: 30
```

```python
# Define customer structure dtype
customer_dtype = np.dtype([('cid', np.str_, 10), ('name', str, 30), ('age', np.int8)])
# Customer records
cust_list =[('C001', 'Alice', 30), ('C002', 'Bob', 40), ('C003', 'Cat', 50)]
# Create customer_dtype ndarray
customers = np.array(cust_list, customer_dtype)
print(customers)
print(customers[0])
print(customers['name'])
print(customers['age'] > 35)
print(customers[customers['age'] > 35])
```

```python
def roll_dice(num_dices, size):
    rng = np.random.default_rng()
# roll the dice
    faces = rng.integers(low=1, high=6, endpoint=True, size=(size,num_dices))
# sum face values
    return np.sum(faces, axis=1)
```

```python
# Roll 1 dice
from IPython import display
# Create empty zero-sized numpy array
face_counts = np.array([], dtype=int)
for i in range(0,100):
    # RoLL 1 dice
    dice_faces = roll_dice(1,100)
    # Concatenate the new result to the previous results
    face_counts = np.concatenate((face_counts, dice_faces), axis=0)
    # Plot histogram of all results
    hist = plt.hist(face_counts)
    # CLear Jupyter display output
    display.clear_output(wait=True)
    # Display the active figure
    display.display(plt.gcf())

display.clear_output(wait=True)
```

```python
#Roll 2 dices

from IPython import display
# Create empty zero-sized numpy array
face_counts = np.array([], dtype=int)
for i in range(0,100):
    # Roll 2 dice
    dice_faces = roll_dice(2,100)
    # Concatenate the new result to the previous results
    face_counts = np.concatenate((face_counts, dice_faces), axis=0)
    # Plot histogram of all results
    hist = plt.hist(face_counts, bins=np.arange(1,14)-0.5, color = 'blue', rwidth= 0.8)
    # CLear Jupyter display output
    display.clear_output(wait=True)
    # Display the active figure
    display.display(plt.gcf())

display.clear_output(wait=True)
```

```python
import numpy as np
mpg_ds = np.genfromtxt(
    'mpg.csv', delimiter=',',names=True, dtype=None, encoding='utf-8',
    converters = {
    3: lambda x : float(x.replace('?','0')),
    7: lambda x: x.replace('1','US').replace('2','EU').replace('3','JP')
             }
                        )
print('shape:', mpg_ds.shape, '\n')
print(mpg_ds[0:10])
```

```python
mpg_dataset = mpg_ds[mpg_ds['horsepower'] > 0]
print(' shape:', mpg_ds.shape)
print(mpg_ds[0:10])
```

```python
#Create a figure with two subplots in one row
fig = plt.subplots(1,2, figsize=(10,4))

origins, counts = np.unique(mpg_ds['origin'], return_counts=True)
print(origins, '\n', counts)

# Subplot 1
plt.subplot(1, 2, 1)  # สร้าง subplot ที่ตำแหน่ง (1, 1)
plt.bar(origins, counts,color = ['red','green','blue'])
plt.title('Car model by origin')
plt.xlabel('Country')
plt.ylabel('Count')
#add annotation
for x,y in np.stack((origins, counts), axis=1):
    y = int(y)
    plt.annotate(y, (x,y), ha ='center',color = 'white', xytext=(x,y-15))
    
# Subplot 2
plt.subplot(1, 2, 2) # สราง subplot ที่ตําแหนง (1, 1)
plt.title('Car model by origin')
plt.pie( counts, labels= origins, autopct ='%1.1f%%',
colors = ['red','green','blue'],
shadow=True, explode=(0.2,0,0)
)
plt.show()
```
