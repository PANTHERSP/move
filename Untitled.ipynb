```python
!pip install python-pptx

from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor

# ฟังก์ชันช่วยในการสร้างข้อความพร้อมปรับแต่งสีและขนาด
def create_text_frame(slide, title_text, body_text):
    title = slide.shapes.title
    title.text = title_text
    title.text_frame.paragraphs[0].font.size = Pt(28)
    title.text_frame.paragraphs[0].font.bold = True
    title.text_frame.paragraphs[0].font.color.rgb = RGBColor(255, 255, 255)
    
    body = slide.shapes.placeholders[1]
    tf = body.text_frame
    tf.clear()  # ลบข้อความตัวอย่างออก
    p = tf.add_paragraph()
    p.text = body_text
    p.font.size = Pt(18)
    p.font.color.rgb = RGBColor(255, 255, 255)
    return tf

# สร้าง PowerPoint ใหม่
prs = Presentation()

# เพิ่มสไลด์แรก (สไลด์หน้าหลัก modern style)
slide_layout = prs.slide_layouts[5]  # สไลด์เปล่า
slide = prs.slides.add_slide(slide_layout)

# เพิ่มรูปร่างสี่เหลี่ยมเต็มพื้นหลังสี modern
background = slide.shapes.add_shape(
    1, Inches(0), Inches(0), prs.slide_width, prs.slide_height
)
background.fill.solid()
background.fill.fore_color.rgb = RGBColor(34, 45, 50)  # สีพื้นหลังเข้ม (modern dark)

# เพิ่มข้อความหัวข้อหน้าหลัก
title_shape = slide.shapes.add_textbox(Inches(0.5), Inches(2), Inches(9), Inches(1))
title_shape.text = "Smart Soil Moisture Monitoring System"
title_frame = title_shape.text_frame
title_frame.paragraphs[0].font.size = Pt(44)
title_frame.paragraphs[0].font.color.rgb = RGBColor(255, 255, 255)
title_frame.paragraphs[0].font.bold = True

# เพิ่มสไลด์สอง (Solution)
slide_layout = prs.slide_layouts[1]  # รูปแบบสไลด์พร้อมข้อความ
slide = prs.slides.add_slide(slide_layout)

background = slide.shapes.add_shape(
    1, Inches(0), Inches(0), prs.slide_width, prs.slide_height
)
background.fill.solid()
background.fill.fore_color.rgb = RGBColor(34, 45, 50)  # สีพื้นหลังเข้ม

create_text_frame(slide, "Solution", 
    "โปรเจคนี้ช่วยเกษตรกรในการปรับปรุงประสิทธิภาพการใช้น้ำ โดยการตรวจสอบความชื้นในดินแบบเรียลไทม์ "
    "ทำให้เกษตรกรสามารถตัดสินใจในการให้น้ำแก่พืชได้อย่างแม่นยำและลดการใช้น้ำโดยไม่จำเป็น")

# เพิ่มสไลด์สาม (Device ที่ใช้)
slide_layout = prs.slide_layouts[1]
slide = prs.slides.add_slide(slide_layout)

background = slide.shapes.add_shape(
    1, Inches(0), Inches(0), prs.slide_width, prs.slide_height
)
background.fill.solid()
background.fill.fore_color.rgb = RGBColor(34, 45, 50)  # สีพื้นหลังเข้ม

create_text_frame(slide, "Devices ที่ใช้", 
    "อุปกรณ์หลักที่ใช้ในโครงการนี้ ได้แก่:\n- เซ็นเซอร์วัดความชื้นในดิน\n- NodeMCU (ESP8266 หรือ ESP32)\n- แบตเตอรี่/แหล่งจ่ายไฟ")

# เพิ่มสไลด์สี่ (Communication)
slide_layout = prs.slide_layouts[1]
slide = prs.slides.add_slide(slide_layout)

background = slide.shapes.add_shape(
    1, Inches(0), Inches(0), prs.slide_width, prs.slide_height
)
background.fill.solid()
background.fill.fore_color.rgb = RGBColor(34, 45, 50)  # สีพื้นหลังเข้ม

create_text_frame(slide, "Communication", 
    "การสื่อสารในระบบนี้ใช้ Wi-Fi ผ่านโปรโตคอล MQTT หรือ HTTP โดย NodeMCU จะส่งข้อมูลความชื้นในดินไปยังคลาวด์หรือเซิร์ฟเวอร์ได้โดยตรง")

# เพิ่มสไลด์ห้า (Visualization)
slide_layout = prs.slide_layouts[1]
slide = prs.slides.add_slide(slide_layout)

background = slide.shapes.add_shape(
    1, Inches(0), Inches(0), prs.slide_width, prs.slide_height
)
background.fill.solid()
background.fill.fore_color.rgb = RGBColor(34, 45, 50)  # สีพื้นหลังเข้ม

create_text_frame(slide, "Visualization", 
    "การแสดงผลข้อมูลสามารถทำได้ผ่าน:\n- แอปพลิเคชันบนมือถือ (iOS/Android)\n- Web Dashboard เช่น Grafana\n- การแจ้งเตือนผ่าน LINE Notify")

# บันทึกไฟล์
prs.save("65011048_ศุภกฤต.pptx")
print("สร้างไฟล์ PowerPoint เสร็จสิ้น: 65011048_ศุภกฤต.pptx")

```
